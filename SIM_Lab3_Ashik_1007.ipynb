# Normal Distribution

A **normal distribution**, also known as a **Gaussian distribution**, is a type of probability distribution that is symmetrical and bell-shaped. In this distribution:

- Most of the data points are clustered around the **mean** (average).
- As you move away from the mean, the probability of the data points decreases, but in a smooth, continuous way.
- The curve is symmetrical, meaning the left side is a mirror image of the right side.

In simpler terms, a normal distribution looks like a bell curve, where most values are near the middle, and fewer values are found as you move further away from the center. Examples include things like people's heights, test scores, or measurement errors.

```python
# Generate 10,000 random values from a normal distribution
SNormalValues = np.random.randn(10000)#mean=0, standard deviation=1

# Create a histogram
plt.hist(SNormalValues, bins=30, edgecolor='black')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

**1.Uniform Random Numbers**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a 1Ã—10 vector of random numbers between 0 and 1
uniform_vector = np.random.rand(10)
print("Uniform Random Vector (1x10):")
print(uniform_vector)

# Plotting histogram
plt.figure(figsize=(6, 4))
plt.hist(uniform_vector, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram of Uniform Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

**2.Normally Distributed Random Numbers**

```python
# Generate 500 random numbers from a standard normal distribution
normal_data = np.random.normal(loc=0, scale=1, size=500)

# Plotting histogram
plt.figure(figsize=(6, 4))
plt.hist(normal_data, bins=30, color='lightgreen', edgecolor='black')
plt.title("Histogram of Normally Distributed Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
